```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("MD.csv")
df.head()
```

```python
df['Genre'] = df['Genre'].str.split(',')
df = df.explode('Genre')
df['Genre'] = df['Genre'].str.strip()

# Group by genre and calculate the average runtime
genre_runtime = df.groupby('Genre')['Runtime (Minutes)'].mean()

# Create a bar chart
plt.figure(figsize=(10, 6))
genre_runtime.plot(kind='bar')
plt.title('Average Runtime by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Runtime (Minutes)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
director_revenue = df.groupby('Director')['Revenue (Millions)'].sum().nlargest(5)
plt.figure(figsize=(8, 8))
plt.pie(director_revenue, labels=director_revenue.index, autopct='%1.1f%%')
plt.title('Market Share of Top 5 Directors by Revenue')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Year', data=df)
plt.title('Number of Movies Released Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Genre'] = df['Genre'].str.split(',')
df = df.explode('Genre')
df['Genre'] = df['Genre'].str.strip()

# Create a box plot
plt.figure(figsize=(12, 6))
sns.boxplot(x='Genre', y='Rating', data=df, palette='Set2')
plt.title('Rating Distribution by Genre')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.scatter(df['Rating'],df['Revenue (Millions)'] )
plt.xlabel('Revenue')
plt.ylabel('Rating')
plt.title('Scatter Plot of Revenue vs Rating')
plt.show()
```
