```python
sales_data={'year':[2010,2012,2015,2020,2022,2024],
      'sales':[10000,50000,80000,40000,100000,200000]}
df=pd.DataFrame(sales_data)
df
```

```python
data = {'subjects': ['Maths', 'Machine Learning', 'Physics', 'Chemistry'],
        'Marks': [75, 80, 60, 55]}
df = pd.DataFrame(data)
df
```

```python
df.plot(y='Marks', kind='pie', labels=df['subjects'], autopct='%1.1f%%', title='Subjects Distribution')
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
X
```

```python
y
```

```python
plt.figure(figsize=(8, 8))
plt.hist(X[:, 3], bins=30, color='purple', alpha=1)
plt.title('Distribution of Petal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.boxplot([X[y==0, 1], X[y==1, 1], X[y==2, 1]], labels=iris.target_names)
plt.title('Sepal Length across Iris Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.grid(True)
plt.show()
```
