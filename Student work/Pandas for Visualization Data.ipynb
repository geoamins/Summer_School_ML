# Line Bar

```python
import pandas as pd
data = {'Year': [2001, 2003, 2005, 2008, 2011],
        'Employees': [20400, 32899, 26807, 39000, 25000]}
df= pd.DataFrame(data)
df.plot(x='Year', y='Employees', title= 'Tesla Coorporate')
```
